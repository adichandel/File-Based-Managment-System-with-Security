```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from fpdf import FPDF
import PyPDF2
import os


class Login():
    def __init__(self,root):
        self.root=root
        self.root.title("File Based Record System")
        self.root.geometry("1350x700+0+0")

        f1=Frame(self.root,bg="grey",bd=5,relief='solid')
        f1.place(x=450,y=150,width=500,height=350)
        self.user=StringVar()
        self.pwd=StringVar()
        Label(f1,text="LOGIN",font=("times new roman",40,"bold"),bg="grey",fg="white").grid(row=0,column=0,columnspan=2)
        Label(f1,text="USERNAME",bg="grey",font=("times new roman",20,"bold")).grid(row=1,column=0,pady=40,padx=20,sticky="w")
        username=Entry(f1,textvariable=self.user,font=("times new roman",15,"bold"),bd=2,relief="solid").grid(row=1,column=1,pady=40,padx=20,sticky="w")
        Label(f1,text="PASSWORD",bg="grey",font=("times new roman",20,"bold")).grid(row=2,column=0,padx=20,sticky="w")
        password=Entry(f1,show="*",textvariable=self.pwd,font=("times new roman",15,"bold"),bd=2,relief="solid").grid(row=2,column=1,padx=20,sticky="w")
        
        button_frame=Frame(f1,bg="grey")
        button_frame.place(x=20,y=250)
        updatebtn=Button(button_frame,command=self.login_fun,text="LOGIN",height=2,width=10).grid(row=0,column=1,padx=30,pady=10)
        deletebtn=Button(button_frame,command=self.reset_fun,text="RESET",height=2,width=10).grid(row=0,column=2,padx=30,pady=10)
        clearbtn=Button(button_frame,command=self.exit_fun,text="EXIT",height=2,width=10).grid(row=0,column=3,padx=30,pady=10)
        
    def login_fun(self):
        if(self.user.get()=="admin" and self.pwd.get()=="admin123"):
            self.root.destroy()
            self.rab(root)
        else:
            messagebox.showerror("ERROR","Invalid Username or Password")
    def reset_fun(self):
        self.user.set("")
        self.pwd.set("")
    def exit_fun(self):
        option=messagebox.askyesno("EXIT","Do you really want to exit")
        if option>0:
            self.root.destroy()

    def rab(self, root):
        self.root = Tk()
        self.root.title("File Based Record System")
        self.root.geometry("1350x700+0+0")
        Label(self.root, text="FILE BASED RECORD SYSTEM", pady=10, font=("times new roman", 40, "bold"), bg="grey",
              fg="white").pack(fill=X)

        self.Student_ID = StringVar()
        self.Student_name = StringVar()
        self.Student_contact = StringVar()
        self.date = StringVar()
        self.Student_course = StringVar()
        self.Student_add = StringVar()
        self.Student_city = StringVar()
        self.degree = StringVar()
        self.pay = StringVar()
        self.id = StringVar()

        student_frame = Frame(self.root, bg="grey", bd=2, relief="solid")
        student_frame.place(x=20, y=100, height=400)
        Label(student_frame, text="Student Details", pady=20, font=("times new roman", 30, "bold"), bg="grey",
              fg="white").grid(row=0, column=0, columnspan=4)
        Label(student_frame, text="Student ID", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=1, column=0)
        s_id = Entry(student_frame, textvariable=self.Student_ID, font=("times new roman", 15, "bold"), bd=2,
                     relief="solid")
        s_id.grid(row=1, column=1, pady=10, padx=20)
        Label(student_frame, text="Student Name", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=2, column=0)
        s_name = Entry(student_frame, textvariable=self.Student_name, font=("times new roman", 15, "bold"), bd=2,
                       relief="solid")
        s_name.grid(row=2, column=1, pady=10, padx=20)
        Label(student_frame, text="Student Contact", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=1, column=2)
        s_contact = Entry(student_frame, textvariable=self.Student_contact, font=("times new roman", 15, "bold"), bd=2,
                          relief="solid")
        s_contact.grid(row=1, column=3, pady=10, padx=20)
        Label(student_frame, text="Date(dd/mm/yyyy)", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=2, column=2)
        s_name = Entry(student_frame, textvariable=self.date, font=("times new roman", 15, "bold"), bd=2,
                       relief="solid")
        s_name.grid(row=2, column=3, pady=10, padx=20)
        Label(student_frame, text="Course", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=3, column=0)
        s_name = Entry(student_frame, textvariable=self.Student_course, font=("times new roman", 15, "bold"), bd=2,
                       relief="solid")
        s_name.grid(row=3, column=1, pady=10, padx=20)
        Label(student_frame, text="Select Degree", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=3, column=2)
        degree = ttk.Combobox(student_frame, textvariable=self.degree, state="readonly",
                              font=("times new roman", 12, "bold"))
        degree['values'] = ("BCA", "MCA", "B.Tech", "M.Tech", "MBA")
        degree.grid(row=3, column=3, padx=20, pady=10)
        Label(student_frame, text="State", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=5, column=0)
        s_address = Entry(student_frame, textvariable=self.Student_add, font=("times new roman", 15, "bold"), bd=2,
                          relief="solid")
        s_address.grid(row=4, column=1, pady=10, padx=20)
        Label(student_frame, text="ID Proof", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=4, column=2)
        degree = ttk.Combobox(student_frame, textvariable=self.id, state="readonly",
                              font=("times new roman", 12, "bold"))
        degree['values'] = ("Adhaar Card", "Student ID Card", "PAN Card")
        degree.grid(row=4, column=3, padx=20, pady=10)
        Label(student_frame, text="City", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=4, column=0)
        s_city = Entry(student_frame, textvariable=self.Student_city, font=("times new roman", 15, "bold"), bd=2,
                       relief="solid")
        s_city.grid(row=5, column=1, pady=10, padx=20)
        Label(student_frame, text="Payment Mode", pady=10, padx=20, font=("times new roman", 20, "bold"), bg="grey",
              fg="white").grid(row=5, column=2)
        pay = ttk.Combobox(student_frame, textvariable=self.pay, state="readonly", font=("times new roman", 12, "bold"))
        pay['values'] = ("Cash", "Internet Banking", "Credit Card")
        pay.grid(row=5, column=3, padx=20, pady=10)

        button_frame = Frame(self.root, bg="grey", bd=2, relief="solid")
        button_frame.place(x=20, y=520)
        save = Button(button_frame,command=self.save, text="SAVE",  width=20, height=2).grid(row=0, column=0, padx=30,
                                                                                             pady=30)
        delete = Button(button_frame, text="DELETE", command=self.delete, width=20, height=2).grid(row=0, column=1,
                                                                                                   padx=30, pady=30)
        clear = Button(button_frame, text="CLEAR", command=self.clear, width=20, height=2).grid(row=0, column=2,
                                                                                                padx=30, pady=30)
        pdf = Button(button_frame, text="SAVE AS PDF", command=self.convert_pdf, width=20, height=2).grid(row=0, column=3, padx=30,
                                                                                             pady=30)
        logout = Button(button_frame, text="LOGOUT", command=self.logout, width=20, height=2).grid(row=0, column=4,
                                                                                                   padx=30, pady=30)
        exit = Button(button_frame, text="EXIT", command=self.exit, width=20, height=2).grid(row=0, column=5, padx=40,
                                                                                             pady=30)
        

        file_frame = Frame(self.root, bg="grey", bd=2, relief="solid")
        file_frame.place(x=1000, y=100, height=400, width=300)
        Label(file_frame, text="ALL FILES", pady=20, font=("times new roman", 20, "bold"), bd=3, relief="groove",
              bg="grey", fg="white").pack(side=TOP, fill=X)
        scroll_y = Scrollbar(file_frame, orient="vertical")
        self.file_list = Listbox(file_frame, yscrollcommand=scroll_y.set)
        scroll_y.pack(side=RIGHT, fill=Y)
        scroll_y.config(command=self.file_list.yview)
        self.file_list.pack(fill="both", expand=1)
        self.file_list.bind("<Button-1>",self.get_data)
        self.show_files()
        self.root.mainloop()

    def save(self):
        present = "no"
        if self.Student_ID.get == "":
            messagebox.showerror("Error", "Student Id required")
        else:
            f = os.listdir("Text Files/")
            if len(f) > 0:
                for i in f:
                    if i.split(".")[0] == self.Student_ID.get():
                        present = "yes"
                if present == "yes":
                    ask = messagebox.askyesno("Update", "File Already present \n DO u want to update it")
                    if ask > 0:
                        self.save_file()
                        messagebox.showinfo("Update", "Record Updated")
                        self.show_files()
                else:
                    self.save_file()
                    messagebox.showinfo("Success", "Record Saved")
                    self.show_files()
            else:
                self.save_file()
                messagebox.showinfo("Success", "Record Saved")
                self.show_files()


        if self.Student_ID.get()=="":
            messagebox.showerror("Error","Student ID must be Requrired")
        else:
            self.save_file()
    def save_file(self):
        f = open(str("Text Files/" + self.Student_ID.get()) + ".txt", "w")
        f.write(
            str(self.Student_ID.get()) + "," +
            str(self.Student_name.get()) + "," +
            str(self.Student_course.get()) + "," +
            str(self.Student_add.get()) + "," +
            str(self.Student_city.get()) + "," +
            str(self.Student_contact.get()) + "," +
            str(self.date.get()) + "," +
            str(self.degree.get()) + "," +
            str(self.id.get()) + "," +
            str(self.pay.get())
        )
        f.close()

        self.show_files()

    def show_files(self):
        files=os.listdir("Text Files/")
        self.file_list.delete(0, END)
        if len(files)>0:
            for i in files:
                self.file_list.insert(END,i)
    def get_data(self,ev):
        get_cursor=self.file_list.curselection()
        f1=open("Text Files/"+self.file_list.get(get_cursor))
        value=[]
        for f in f1:
            value=f.split(",")
        #print(value)
        self.Student_ID.set(value[0])
        self.Student_name.set(value[1])
        self.Student_contact.set(value[5])
        self.date.set(value[6])
        self.Student_course.set(value[2])
        self.Student_add.set(value[3])
        self.Student_city.set(value[4])
        self.degree.set(value[7])
        self.pay.set(value[9])
        self.id.set(value[8])


        #print(self.file_list.get(get_cursor))



    def delete(self):
        present="no"
        if self.Student_ID.get=="":
            messagebox.showerror("Error","Student Id required")
        else:
            f=os.listdir("Text Files/")
            if len(f)>0:
                for i in f:
                    if i.split(".")[0]==self.Student_ID.get():
                        present="yes"
                if present=="yes":
                    ask=messagebox.askyesno("Delete","Do u really want to delele")
                    if ask>0:
                        os.remove("Text Files/"+self.Student_ID.get()+".txt")
                        messagebox.showinfo("Suceess","Deleted Successfully")
                        self.show_files()
            else:
                messagebox.showerror("Error","File not found")



    def clear(self):
        self.Student_ID.set("")
        self.Student_name.set("")
        self.Student_contact.set("")
        self.date.set("")
        self.Student_course.set("")
        self.Student_add.set("")
        self.Student_city.set("")
        self.degree.set("")
        self.pay.set("")
        self.id.set("")


    def logout(self):
        self.root.destroy()
        root=Tk()
        ob=Login(root)
        root.mainloop()






    def exit(self):
        ask = messagebox.askyesno("Exit", "Do u really want to Exit")
        if ask > 0:
            self.root.destroy()

    def convert_pdf(self):
        pdf=FPDF()
        pdf.add_page()
        get_cursor=self.file_list.curselection()
        f=open("Text Files/"+self.file_list.get(get_cursor))
        pdf.set_font("Arial", size = 30) 
        pdf.cell(200,20,txt="Student Details",ln=1,align="C")
        l=''
        for x in f:
            l=l+x
        x=[]
        x=l.split(",")
        pdf.set_font("Arial", size = 15) 
        pdf.cell(200, 10, txt ="Student ID: "+x[0], ln = 2, align = 'L') 
        pdf.cell(200, 10, txt ="Name: "+ x[1], ln = 3, align = 'L') 
        pdf.cell(200, 10, txt ="Course: "+ x[2], ln = 4, align = 'L') 
        pdf.cell(200, 10, txt ="Degree: "+ x[7], ln = 5, align = 'L') 
        pdf.cell(200, 10, txt ="City: "+ x[3], ln = 6, align = 'L') 
        pdf.cell(200, 10, txt ="State: "+ x[4], ln = 7, align = 'L') 
        pdf.cell(200, 10, txt ="Phone Number: "+ x[5], ln = 8, align = 'L') 
        pdf.cell(200, 10, txt ="Date Of Birth: "+ x[6], ln = 9, align = 'L') 
        pdf.cell(200, 10, txt ="Identity Proof: "+ x[8], ln = 10, align = 'L')
        pdf.cell(200, 10, txt ="Mode Of Payment: "+ x[9], ln = 11, align = 'L') 
        pdf.output("PDF Files/"+x[0]+".pdf")  
        pdfFile = open("PDF Files/"+x[0]+".pdf", 'rb')
        pdfReader = PyPDF2.PdfFileReader(pdfFile)
        pdfWriter = PyPDF2.PdfFileWriter()
        for pageNum in range(pdfReader.numPages):
            pdfWriter.addPage(pdfReader.getPage(pageNum))
        pdfWriter.encrypt(x[5])
        resultPdf = open("PDF Files/"+x[0]+".pdf", 'wb')
        pdfWriter.write(resultPdf)
        resultPdf.close()
        messagebox.showinfo("Successfull","File successfully saved in PDF format")

root = Tk()
ob = Login(root)
root.mainloop()
```
